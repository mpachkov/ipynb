```python
x = np.linspace(0, 20, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

pylab.plot(x, y1, '.b', label='sine')
pylab.plot(x, y2, '--r', label='cosine')
pylab.legend(loc='upper right')
pylab.ylim(-1.5, 2.0)
```

## Object-oriented approach
* matplotlib created a Figure instance, which is an object describing the plot window and its properties, and containing lists of all its elements
* matplotlib created an Axes element within the figure. An axes can be thought of as a plotting window, where data can be arranged by x and y coordinates.

```python
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 1000)
y = np.sin(x)
y2 = np.cos(x)
fig = plt.figure()  # a new figure window
ax = fig.add_subplot(1, 1, 1)  # specify (nrows, ncols, axnum)
ax.plot(x, y)
ax.plot(x, y2)
ax.set_ylim(-1.5, 2.0)
ax.legend(['sine', 'cosine'])
ax.grid()
```

### Two plots in one figure

```python
fig = plt.figure()  # a new figure window
ax1 = fig.add_subplot(2, 1, 1)  # specify (nrows, ncols, axnum)
ax2 = fig.add_subplot(2, 1, 2)
ax1.plot(x, y)
ax1.set_title("Sine")
ax2.plot(x, y2)
ax2.set_title("Cosine")
#fig.subplots_adjust(wspace=0.8, hspace=0.3)  # just subplot spacing
```

### LaTeX in matplotlib

```python
fig = plt.figure()  # a new figure window
ax1 = fig.add_subplot(1, 2, 1)  # specify (nrows, ncols, axnum)
ax2 = fig.add_subplot(1, 2, 2)
ax1.plot(x, y)
ax1.set_title("$\sin(x)$")
ax2.plot(x, y2)
ax2.set_title("$\cos(x)$")
```

### Adding text to figures

```python
fig = plt.figure()  # a new figure window
ax = fig.add_subplot(1, 1, 1)  # specify (nrows, ncols, axnum)
ax.plot(x, y)
ax.plot(x, y2)
ax.set_ylim(-1.5, 2.0)
ax.legend(['sine', 'cosine'])
ax.text(3, 1, '$\sin(x)$', size=16, ha='center', va='center')
ax.text(1.5, -0.8, '$\cos(x)$ afadg ete!', size=16, ha='center', va='center')
ax.grid()
```

### Custom tick labels, grid

```python
x = [1, 2, 3, 4]
x_labels = ['sample1', 'sample2', 'sample3', 'sample4']
y = [1, 1.5, 2.8, 2.2]
fig = plt.figure()  # a new figure window
ax = fig.add_subplot(1, 1, 1)  # specify (nrows, ncols, axnum)
ax.plot(x, y, 'ro')
# ajust our view
ax.set_ylim(0, 3)
ax.set_xlim(0, 5)
ax.set_xticks(x)
ax.set_xticklabels(x_labels, rotation=-90)
ax.grid()
fig.tight_layout()
# Save the figure
fig.savefig("figure.pdf")
```

```python
# auto-adjust figure to avoid elements clipping
fig.tight_layout()
# Save the figure
fig.savefig("figure.pdf")
```

### Histogram, boxplot, errorbars

```python
import numpy.random
x = numpy.random.normal(size=10000)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist(x, bins=5, normed=True)
ax.set_title("Normal distribution")
```

```python
gaussian_numbers = numpy.random.normal(size=1000)
uniform_numbers = numpy.random.uniform(low=-3, high=3, size=1000)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist(gaussian_numbers, bins=20, histtype='stepfilled',
        normed=True, color='b', label='Gaussian')
ax.hist(uniform_numbers, bins=20, histtype='stepfilled',
        normed=True, color='r', alpha=0.5, label='Uniform')
ax.set_title("Gaussian/Uniform Histogram")
ax.set_xlabel("X")
ax.set_ylabel("Probability")
ax.legend()
```

```python
x = numpy.random.normal(size=10000)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.hist(x, bins=30, histtype='stepfilled', cumulative=True)
```

```python
x = numpy.random.normal(size=10000)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.boxplot(x)
```

```python
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)
yerr = 0.1 + 0.2*np.sqrt(x)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#ax.errorbar(x, y, yerr=yerr)
#ax.errorbar(x, y, yerr=yerr, fmt='-o')
ax.errorbar(x, y, yerr=yerr, ecolor="r", fmt="-o")
```

## 3D fun

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

c = x + y

ax.scatter(x, y, z, c=c)
```
